```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import re

BASE = Path("results/cnn_cifar10_iid_lrate")
ALGOS = ["Algorithm1", "FedAvg"]

def load_algo(algo):
    rows = []
    for f in sorted((BASE / algo).glob("*.csv")):
        df = pd.read_csv(f)
        df["algo"] = algo
        rows.append(df)
    return pd.concat(rows, ignore_index=True)

data = pd.concat([load_algo(a) for a in ALGOS], ignore_index=True)

for algo, gdf in data.groupby("algo"):
    clusters = sorted(gdf["data_n_clusters"].unique())
    lrates   = sorted(gdf["algo_lrate"].unique())

    fig, axes = plt.subplots(1, len(clusters), figsize=(5 * len(clusters), 4), sharey=True)
    if len(clusters) == 1:
        axes = [axes]

    for ax, nc in zip(axes, clusters):
        sub = gdf[gdf["data_n_clusters"] == nc]
        for lr in lrates:
            curve = sub[sub["algo_lrate"] == lr].sort_values("iter")
            ax.plot(curve["iter"], curve["accuracy"], label=f"lr={lr}")
        ax.set_title(f"n_clusters={nc}")
        ax.set_xlabel("Round")
        ax.legend(fontsize=8)

    axes[0].set_ylabel("Accuracy")
    fig.suptitle(algo, fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.show()
```
